# Li et al. (2017): *"The Market for Electric Vehicles: Indirect Network Effects and Policy Design"*

*Settings:*

```python
# Parameters
β1 = 0.8
β2 = -1.5
β3 = 1
γ1 = 0.4
γ2 = 1
δ = 0.9
# Time periods
T = 80
t_grid = np.linspace(0,T,T+1).astype(int)
# Variables
p_grid, x_grid, z_grid = np.zeros(T+1),np.zeros(T+1),np.zeros(T+1)
p_grid[:] = 30000
x_grid[:] = 16
z_grid[:] = 2
```

*Transition equation:*

```python
def constant(β1,γ2,z,β2,p,β3,x):
    return β1*γ2*z+β2*np.log(p)+β3*x
def excess_EV_demand(Q,*params):
    β1,β2,β3,γ1,γ2,δ,p,z,x,Q_minus = params
    c = constant(β1,γ2,z,β2,p,β3,x)
    return Q-δ*Q_minus - Q**(β1*γ1)*np.exp(c)
def min_excess_EV_demand(Q,*params):
    return (excess_EV_demand(Q,*params))**2
```

*Critical mass value:*

```python
def Q_bar(β1,γ2,z,β2,p,β3,x,γ1):
    c = constant(β1,γ2,z,β2,p,β3,x)
    Q1 = np.exp(c/(1-γ1*β1))
    return (γ1*β1)**(1/(1-γ1*β1))*Q1
```

*Steady state values:*

```python
def Q_steady_state(β1,γ2,z,β2,p,β3,x):
    c = constant(β1,γ2,z,β2,p,β3,x)
    return (np.exp(c)/(1-δ))**(1/(1-γ1*β1))
def q_steady_state(Q_st,δ):
    return Q_st*(1-δ)
```

*Algorithm for solving model:*

```python
def solve_model(β1,β2,β3,γ1,γ2,δ,T,Q0):
   
    # Vectors:
    t_grid = np.linspace(0,T,T+1).astype(int)
    Q_sol = np.zeros(T+1)
    
    # Initial condition
    Q_sol[0] = Q0
    
    # Critical mass condition:
    Q̄ = [Q_bar(β1,γ2,z_grid[t],β2,p_grid[t],β3,x_grid[t],γ1) for t in t_grid]
    
    # Steady state values:
    Q_st = Q_steady_state(β1,γ2,z_grid[T],β2,p_grid[T],β3,x_grid[T])
    q_st = q_steady_state(Q_st,δ)
    
    for t in t_grid[1:]:
        if Q_sol[t-1]<=Q̄[t-1]:
            Q_sol[t] = δ*Q_sol[t-1]
        else:
            Q_sol[t] = fsolve(excess_EV_demand,x0=np.array([Q_sol[t-1]]),args=(β1,β2,β3,γ1,γ2,δ,p_grid[t],z_grid[t],x_grid[t],Q_sol[t-1]))[0]
    d_out = {}
    d_out['Q'] = pd.Series(Q_sol,index=pd.Index(t_grid,name='t'),name='Q')
    d_out['Q̄'] = pd.Series(Q̄,index=pd.Index(t_grid,name='t'),name='Q̄')
    d_out['q'] = (d_out['Q'] - δ*d_out['Q'].shift()).dropna().rename('q')
    d_out['Q_st'] = Q_st
    d_out['q_st'] = q_st
    return d_out
```

*Plot Excess EV demand in first period:*

```python
Q̄ = Q_bar(β1,γ2,z_grid[0],β2,p_grid[0],β3,x_grid[0],γ1)
Q_st = Q_steady_state(β1,γ2,z_grid[0],β2,p_grid[0],β3,x_grid[0])
Q0_opt = fsolve(excess_EV_demand,x0=np.array([Q_st]),args=(β1,β2,β3,γ1,γ2,δ,p_grid[1],z_grid[1],x_grid[1],Q̄))[0]
Q1_opt = fsolve(excess_EV_demand,x0=np.array([Q_st]),args=(β1,β2,β3,γ1,γ2,δ,p_grid[1],z_grid[1],x_grid[1],Q0_opt*2))[0]
N_Q = 99
ϵ = np.finfo(float).eps
Q0_grid = np.concatenate((
    np.linspace(0,Q̄-ϵ,N_Q),
    np.array([Q̄]),
    np.linspace(Q̄+ϵ,Q0_opt-ϵ,N_Q),
    np.array([Q0_opt]),
    np.linspace(Q0_opt+ϵ,Q0_opt*2,N_Q),
))
Q1_grid = np.concatenate((
    np.linspace(0,Q̄-ϵ,N_Q),
    np.array([Q̄]),
    np.linspace(Q̄+ϵ,Q1_opt-ϵ,N_Q*2),
    np.array([Q1_opt]),
))
F_val = np.empty((len(Q0_grid),len(Q1_grid)))
F_zero = np.zeros(len(Q1_grid))
Q_zero = F_zero.copy()
for i in range(len(Q0_grid)):
    Q_minus = Q0_grid[i]
    Q_zero[i] = fsolve(excess_EV_demand,x0=np.array([Q_st]),args=(β1,β2,β3,γ1,γ2,δ,p_grid[1],z_grid[1],x_grid[1],Q_minus))[0]
    F_zero[i] = excess_EV_demand(Q_zero[i],*(β1,β2,β3,γ1,γ2,δ,p_grid[1],z_grid[1],x_grid[1],Q_minus))
    for j in range(len(Q1_grid)):
        Q_plus = Q1_grid[j]
        F_val[i,j] = excess_EV_demand(Q_plus,*(β1,β2,β3,γ1,γ2,δ,p_grid[1],z_grid[1],x_grid[1],Q_minus))
```
